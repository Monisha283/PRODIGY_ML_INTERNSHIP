# **Linear Regression Model**

```python
# Step 1 : import library
import pandas as pd
```

```python
# Step 2 : import data
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
df.shape
```

```python
y = df['MEDV']
X = df.drop(['MEDV'],axis=1)
```

```python
y.shape
```

```python
# Step 4 : train test split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=2529)
X_train
```

```python
# check shape of train and test sample
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Step 5 : select model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# Step 6 : train or fit model
model.fit(X_train,y_train)
```

```python
model.intercept_ # this give intercept point value of best fit line
```

```python
model.coef_ # this give as the slope of best fit lineline
```

```python
# Step 7 : predict model
y_pred = model.predict(X_test)
```

```python
y_pred # predict all value
```

```python
# Step 8 : model accuracy
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```

**16% error is show\
than 100-16 = 84%\
model is 84% accurate**

```python
mean_squared_error(y_test,y_pred)
```

# **Visualization of Actual Vs predicted result**

```python
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual Vs Predicted")
plt.show()
```
